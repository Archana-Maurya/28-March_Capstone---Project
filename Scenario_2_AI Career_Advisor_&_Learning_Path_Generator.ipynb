```python
# 🔷 Scenario 2: AI Career Advisor & Learning Path Generator

# 🎯 Objective:
# Build a multi-step intelligent advisor system using agent workflows.

# 💼 Problem Statement
# Students need guidance on career paths based on their skills.

# Your system should:

# Analyze student profile
# Suggest career roles
# Generate learning roadmap

# 👥 Agents to Build

# Profile Analyzer Agent
# Understands student background

# Career Recommendation Agent
# Suggests suitable roles

# Skill Gap Agent
# Identifies missing skills

# Learning Path Agent
# Creates step-by-step roadmap

# ⚙️ Tasks to Implement

# Sequential workflow (LangGraph preferred)
# State passing between agents
# Conditional logic (beginner vs intermediate)

# 💡 Sample Input
# “I know Python basics and statistics, and I want to become a Data Scientist”

# 🧪 Expected Output

# Suggested roles (e.g., Data Scientist, Analyst)
# Skill gaps (ML, SQL, etc.)
# Learning roadmap (step-by-step)

# 🧰 Tools (Choose Any)

# LangGraph (recommended)
# CrewAI
# AutoGen
# Python logic-based workflow
# Microsoft Copilot Studio

# 📦 Deliverables

# Workflow implementation
# Output for at least 2 student profiles
# Diagram of agent flow






#            ┌──────────────────────┐
#            │     User Input       │
#            │ (Student Profile)    │
#            └─────────┬────────────┘
#                      ↓
#         ┌──────────────────────────┐
#         │ Profile Analyzer Agent   │
#         │ - Extract skills         │
#         │ - Detect level           │
#         └─────────┬────────────────┘
#                   ↓
#         ┌──────────────────────────┐
#         │ Career Recommendation    │
#         │ Agent                    │
#         │ - Suggest roles          │
#         └─────────┬────────────────┘
#                   ↓
#         ┌──────────────────────────┐
#         │ Skill Gap Agent          │
#         │ - Identify missing skills│
#         └─────────┬────────────────┘
#                   ↓
#         ┌──────────────────────────┐
#         │ Learning Path Agent      │
#         │ - Generate roadmap       │
#         └─────────┬────────────────┘
#                   ↓
#         ┌──────────────────────────┐
#         │ Final Output Generator   │
#         │ - Structured response    │
#         └──────────────────────────┘


# ================================
# 🔥 INSTALL
# ================================
!pip install openai

# ================================
# 🔐 SECURE API SETUP
# ================================
import os
from openai import OpenAI

# 👉 Colab + Environment दोनों support
try:
    from google.colab import userdata
    api_key = userdata.get("GROQ_API_KEY")
except:
    api_key = None

api_key = api_key or os.getenv("GROQ_API_KEY")

if not api_key:
    raise ValueError("❌ API key not found. Set GROQ_API_KEY securely.")

client = OpenAI(
    api_key=api_key,
    base_url="https://api.groq.com/openai/v1"
)

MODEL = "llama-3.3-70b-versatile"

# ================================
# 🧠 STATE CLASS
# ================================
class State:
    def __init__(self, user_input):
        self.user_input = user_input
        self.profile = ""
        self.roles = ""
        self.skill_gap = ""
        self.roadmap = ""

# ================================
# 🔗 LLM HELPER (SAFE)
# ================================
def llm_call(prompt):
    try:
        res = client.chat.completions.create(
            model=MODEL,
            messages=[{"role": "user", "content": prompt}]
        )
        return res.choices[0].message.content.strip()
    except Exception as e:
        return f"⚠️ Error: {str(e)}"

# ================================
# 🤖 AGENTS
# ================================

def profile_analyzer(state):
    prompt = f"""
You are a Profile Analyzer Agent.

Analyze the student profile:
{state.user_input}

Return:
- Skill Level (Beginner / Intermediate)
- Key Skills
- Interests
"""
    state.profile = llm_call(prompt)
    return state


def career_agent(state):
    prompt = f"""
You are a Career Recommendation Agent.

Based on profile:
{state.profile}

Suggest 2-3 suitable career roles with short explanation.
"""
    state.roles = llm_call(prompt)
    return state


def skill_gap_agent(state):
    prompt = f"""
You are a Skill Gap Agent.

Given career roles:
{state.roles}

Identify missing skills required to achieve these roles.
"""
    state.skill_gap = llm_call(prompt)
    return state


def learning_path_agent(state):
    level = "beginner" if "beginner" in state.profile.lower() else "intermediate"

    prompt = f"""
You are a Learning Path Agent.

Create a step-by-step roadmap for a {level} student.

Focus on:
- Skills to learn
- Tools/technologies
- Suggested order

Missing skills:
{state.skill_gap}
"""
    state.roadmap = llm_call(prompt)
    return state

# ================================
# 🔄 WORKFLOW ENGINE
# ================================

def run_system(user_input):
    state = State(user_input)

    state = profile_analyzer(state)
    state = career_agent(state)
    state = skill_gap_agent(state)
    state = learning_path_agent(state)

    return f"""
👤 Profile Analysis:
{state.profile}

💼 Career Suggestions:
{state.roles}

📉 Skill Gap:
{state.skill_gap}

📚 Learning Roadmap:
{state.roadmap}
"""

# ================================
# 🧪 TEST CASES
# ================================

inputs = [
    "I know Python basics and statistics, want to become Data Scientist",
    "I am beginner in coding and interested in web development"
]

for i in inputs:
    print("USER:", i)
    print(run_system(i))
    print("="*60)
```
